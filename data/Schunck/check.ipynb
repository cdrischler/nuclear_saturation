# Investigating the data

Let's take a look at the data provided by Mike Grosskopf:

```python
import pandas as pd
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, tight_layout=True)
for i in range(1,4):
    dd = pd.read_csv(f"./samples{i}.csv", names=["rho0","E/A"])
    plt.scatter(dd["rho0"], dd["E/A"], label=f"Config {i}")
ax.legend()
ax.set_xlabel("$n_0$")
ax.set_ylabel("$E_0/A$")
```

Did Mike generate the data with a lower cutoff of 0.155? It seems the distribution (i.e., the samples) are chopped off at that density.
